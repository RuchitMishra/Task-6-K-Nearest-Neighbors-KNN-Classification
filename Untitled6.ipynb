```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
```

```python
#prepairing the data

df=pd.read_csv('/content/Iris.csv')
df.head()
df.drop('Id',axis=1,inplace=True)
#features andd target
x=df.drop('Species',axis=1)
y=df['Species']

#narmalize feature
scaler=StandardScaler()
x_scaler=scaler.fit_transform(x)

#encoding
le=LabelEncoder()
y_encoder=le.fit_transform(y)

#split data into training and testing sets
X_train,X_test,y_train,y_test=train_test_split(x_scaler,y_encoder,test_size=0.2,random_state=42)
#feature scaling
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)


```

```python
#2.Use KNeighborsClassifier from sklearn.
#set n_neighbour to a value less than or equalto the smallest class size
#for example, set it to 3 or less based on the error message
knn=KNeighborsClassifier(n_neighbors=27)#3.hear at place of n_neighbors we can put different value of k
knn.fit(X_train,y_train)                # till 27 we will get 100% accuracy
knn_pred=knn.predict(X_test)
knn_accuracy=accuracy_score(y_test,knn_pred)
knn_confusion_matrix=confusion_matrix(y_test,knn_pred)
```

```python
#4.Evaluate model using accuracy, confusion matrix.
print(f'K-Nearest Neighbors Accuracy:{knn_accuracy*100:.2f}')
print('K-Nearest Neighbors Confusion Matrix:')
print(knn_confusion_matrix)
```

```python
# 5.Visualize decision boundaries.
feature_pair = [2, 3]
X_vis = x_scaler[:, feature_pair]

knn_vis = KNeighborsClassifier(n_neighbors=3)
knn_vis.fit(X_vis, y_encoder)

h = 0.02
x_min, x_max = X_vis[:, 0].min() - 1, X_vis[:, 0].max() + 1
y_min, y_max = X_vis[:, 1].min() - 1, X_vis[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = knn_vis.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.Pastel2)
plt.scatter(X_vis[:, 0], X_vis[:, 1], c=y_encoder, edgecolor='k', cmap=plt.cm.Set1)
plt.xlabel("Petal Length (standardized)")
plt.ylabel("Petal Width (standardized)")
plt.title(f"KNN Decision Boundary (K={3})")
plt.show()
```
